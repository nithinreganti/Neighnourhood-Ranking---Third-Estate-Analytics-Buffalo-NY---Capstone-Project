```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
Tows=pd.read_csv(r'F:\Gentrification\Tows.csv')
```

```python
Tows.head()
```

```python
Tows['TOW DATE']=Tows['TOW DATE'].apply(lambda x:x[6] + x[7] +x[8] +x[9]) 
Tows['TOW DATE']=Tows['TOW DATE'].apply(lambda x:float(x))
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
Tows['TOW DESCRIPTION'] = lb_make.fit_transform(Tows['TOW DESCRIPTION'])
Tows['NEIGHBORHOOD'] = lb_make.fit_transform(Tows['NEIGHBORHOOD'].astype(str))
Tows['COUNCIL DISTRICT'] = lb_make.fit_transform(Tows['COUNCIL DISTRICT'].astype(str))
Tows['POLICE DISTRICT'] = lb_make.fit_transform(Tows['POLICE DISTRICT'].astype(str))
Tows['CENSUS TRACT'] = lb_make.fit_transform(Tows['CENSUS TRACT'].astype(str))


```

```python
feature_cols =['TOW DATE','TOW DESCRIPTION','CENSUS TRACT']
```

```python
 Tows['RANK'].fillna((Tows['RANK'].mean()),inplace=True)
```

```python
# Tows['RANK'].fillna(0)
```

```python
Tows["RANK"].astype('category')
```

```python
X = Tows[feature_cols] # Features
y = Tows.RANK
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# clf = DecisionTreeClassifier()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg=DecisionTreeRegressor( max_depth=6)
```

```python
reg = reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
Predicted_rank=[] 
Predicted_rank= np.round(y_pred)
```

```python
reg.score(X_test, y_test)
```

```python
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df.head(5)
```

```python
# from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=20, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred,})
df
```
